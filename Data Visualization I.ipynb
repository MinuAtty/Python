# Matplotlib

-  Matplotlib is an amazing visualization library in Python for 2D plots of arrays.


- Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work   with the broader SciPy stack.  


-  One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible  visuals.  


-  Matplotlib consists of several plots like line, bar, scatter, histogram etc.

## Two ways to create Matplotlib plot

### 1. Functional Approach: Using the basic matplotlib command, we can easily create a plot.

````python
# Plotting a line graph of 'total_cases' against 'time line'
plt.plot(slcd['time line'], slcd['total_cases'])
````

### 2. Object oriented Interface: This is the best way to create plots. The concept is create Figure objects and call methods off it. Creating a blank Figure the .figure() method 

````python
# Creating a figure and an axes for plotting 'total_cases' against 'time line'
fig = plt.figure()
ax = fig.add_axes([0,0,3,1]) # Setting the position and size of the axes
ax.plot(slcd['time line'], slcd['total_cases'])
````

````python
# Creating a figure and an axes for plotting 'total_deaths' against 'time line'
fig = plt.figure()
ax = fig.add_axes([0,0,3,1]) # Setting the position and size of the axes
ax.plot(slcd['time line'], slcd['total_deaths'])
````

````python
# Creating a figure and an axes for plotting 'total_cases' against 'time line' with labels and legend
fig, ax = plt.subplots() # Creating a figure and an axes
ax.plot(slcd['time line'], slcd['total_cases'], label='total cases') # Plotting data and setting label
ax.set_xlabel('Time line') # Setting x-axis label
ax.set_ylabel('Total cases') # Setting y-axis label
ax.set_title('Covid cumulative') # Setting title
ax.legend(loc='best') # Adding legend with the best position
# plt.savefig('covid data.png', dpi = 300)
````

````python
# Creating a figure and an axes for plotting 'total_deaths' against 'time line' with labels and legend
fig, ax = plt.subplots() # Creating a figure and an axes
ax.plot(slcd['time line'], slcd['total_deaths'], label='total deaths') # Plotting data and setting label
ax.set_xlabel('Time line') # Setting x-axis label
ax.set_ylabel('Total deaths') # Setting y-axis label
ax.set_title('Covid cumulative') # Setting title
ax.legend(loc='best') # Adding legend with the best position
# plt.savefig('covid data.png', dpi = 300)
````

## Sub plots

Matplotlibâ€™s pyplot API has a convenience function called subplots() which acts as a utility wrapper and helps in creating common layouts of subplots, including the enclosing figure object, in a single call.

 - specifying rows and columns
 - accessign a specific axes and plot
 - sharing axis across rows and columns
 - figure layout options
 - grid layout to place subplots within a figure

````python
# Creating a figure with 1 row and 2 columns for plotting multiple graphs
fig, a = plt.subplots(1, 2) # Creating a figure with 1 row and 2 columns
fig.tight_layout(pad=2) # Adjusting spacing between subplots
````

````python
# Creating a figure with 2 row and 1 columns for plotting multiple graphs
fig, a = plt.subplots(2, 1) # Creating a figure with 2 row and 1 columns
fig.tight_layout(pad=2) # Adjusting spacing between subplots
````

## Insets

 - one of the useful things this allows you to do is include "inset" figures which are often used to show greater detail of a region of the enclosing plot
 
 - allows easily placing insets in the corners of the axes by specifying a width and height and optionally a location (loc) that accepts locations as codes, similar to legend.
 
 - specified interms of actual sizes or relative size as percentages of parent figure size.

````python
# Importing inset_axes function for adding inset axes
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
````

````python
# Create a figure with two subplots arranged horizontally (1 row, 2 columns)
# Set the figure size to 3 times the default width and 3 inches height
fig, (ax, ax2) = plt.subplots(1, 2, figsize=[3*2, 3])

# Adjust the layout of the subplots, adding padding between them
fig.tight_layout(pad=2)

# Create inset axes inside the main axes 'ax' with a width of 1.3 times and height of 0.9 times the size of 'ax'
axins = inset_axes(ax, width=1.3, height=0.9)

# Create inset axes inside the main axes 'ax2' with a width of 30% and height of 40% of 'ax2' size
# Locate the inset in the lower right corner (loc=4)
axins2 = inset_axes(ax2, width="30%", height="40%", loc=4)

# Turn off tick labels for the inset axes 'axins'
axins.tick_params(labelleft=False, labelbottom=False)
````

````python
# Create another figure with two subplots arranged horizontally (1 row, 2 columns)
# Set the figure size to 5.5 inches width and 2.8 inches height
fig, (ax, ax2) = plt.subplots(1, 2, figsize=[5.5, 2.8])

# Create inset axes inside the main axes 'ax' with a width of 1.3 times and height of 0.9 times the size of 'ax'
axins = inset_axes(ax, width=1.3, height=0.9)

# Create inset axes inside the main axes 'ax' with a width of 30% and height of 40% of 'ax' size
# Locate the inset in the upper left corner (loc=2)
axins2 = inset_axes(ax, width="30%", height="40%", loc=2)

# Create inset axes inside the main axes 'ax2' with a width of 30% and height of 1 times the size of 'ax2'
# Locate the inset in the upper left corner (loc=2) with border padding of 1
axins3 = inset_axes(ax2, width="30%", height=1., loc=2, borderpad=1)

# Create inset axes inside the main axes 'ax2' with a width of 20% and height of 20% of 'ax2' size
# Locate the inset in the lower right corner (loc=4) with border padding of 1
axins4 = inset_axes(ax2, width="20%", height="20%", loc=4, borderpad=1)
````

## Turning tick parameters off

````python
# Create a figure with two subplots arranged horizontally (1 row, 2 columns)
# Set the figure size to 10.5 inches width and 5.8 inches height
fig, (ax, ax2) = plt.subplots(1, 2, figsize=[10.5, 5.8])

# Create inset axes inside the main axes 'ax' with a width of 1.3 times and height of 0.9 times the size of 'ax'
axins = inset_axes(ax, width=1.3, height=0.9) # at the default upper right location

# Create inset axes inside the main axes 'ax' with a width of 30% and height of 40% of 'ax' size
# Locate the inset in the lower right corner (loc=4) with border padding of 2
axins2 = inset_axes(ax, width="30%", height="40%", loc=4, borderpad=2) 
# loc 1 = upper left
#loc 2 =  upper right
#loc 3 =  lower left
# loc4 = lower right

# Create inset axes inside the main axes 'ax2' with a width of 30% and height of 1 times the size of 'ax2'
# Locate the inset in the upper left corner (loc=2)
axins3 = inset_axes(ax2, width="30%", height=1., loc=2)

# Create an inset in the lower right corner (loc=4) of the 'ax2' subplot with width of 20% and height of 20% of 'ax2' size
# Set the border padding to 2
axins4 = inset_axes(ax2, width="20%", height="20%", loc=4, borderpad=2)

# Turn off tick labels for the inset axes 'axins', 'axins2', and 'axins3'
for axi in [axins, axins2, axins3]:
    axi.tick_params(labelleft=False, labelbottom=False)

# Set the x-axis label for 'axins2' inset axes
axins2.set_xlabel('xlabel')
````

````python
# Read the CSV file 'owid-covid-data.csv' into a DataFrame named 'country_data'
# The file path is specified using a raw string literal (r"datasets\owid-covid-data.csv")
country_data = pd.read_csv(r"datasets\owid-covid-data.csv")
````

````python
# Display information about the DataFrame 'country_data', such as column names, data types, and memory usage
country_data.info()
````

````python
# Group the 'country_data' DataFrame by the 'date' column and aggregate the 'total_cases' and 'total_deaths' columns by summing them up
# Create a new DataFrame named 'daily_data' containing the total cases and total deaths aggregated on each date
daily_data = country_data.groupby('date').agg({'total_cases': sum, 'total_deaths': sum})
````

````python
daily_data
````

````python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Create a figure and a single subplot (1 row, 1 column) with the specified figure size
fig, ax = plt.subplots(1, 1, figsize=[9, 6])

# Create an inset axes inside the main axes 'ax' with a width of 50% and height of 50% of 'ax' size
# Locate the inset in the upper left corner (loc=2) with border padding of 3
axins = inset_axes(ax, width="50%", height="50%", loc=2, borderpad=3)

# Define the x and y variables for the scatter plot from the 'country_data' DataFrame
x = country_data['total_cases']
y = country_data['total_tests']

# Create two new variables 'obj1' and 'obj2' by multiplying 'x' and 'y' by 100 respectively
obj1 = 100 * x
obj2 = 100 * y

# Create a scatter plot on the inset axes 'axins' with 'x' and 'y' as coordinates
# Color the objects using the 'obj1' variable, set the size using the 'obj2' variable, and set the opacity to 0.2
# Use the 'Blues' colormap for coloring
axins.scatter(x, y, c=obj1, s=obj2, alpha=0.2, cmap='Blues')

# Set x-label, y-label, and title for the inset axes 'axins'
axins.set_xlabel('total_tests')
axins.set_ylabel('total_cases')
axins.set_title('Test Vs Detections')

# Plot the 'total_cases' and 'total_deaths' from the 'daily_data' DataFrame on the main axes 'ax'
ax.plot(daily_data['total_cases'], label='total cases', color='red')
ax.plot(daily_data['total_deaths'], label='total deaths', color='blue', linestyle='-.')

# Turn off tick labels for the inset axes 'axins'
for axi in [axins]:
    axi.tick_params(labelleft=False, labelbottom=False)
````

````python
# Grouped bar charts
positions = list(range(len(data['Country'])))
width = 0.25 

# Correcting for potential space in column names
data.columns = data.columns.str.strip()

# Plotting the bars again with corrected column names
fig, ax = plt.subplots(figsize=(10, 5))

# Create bars for Gold medals
plt.bar(positions, data['Gold'], width, alpha=0.7, label='Gold')

# Create bars for Silver medals, with positions offset by width
plt.bar([p + width for p in positions], data['Silver'], width, alpha=0.7, label='Silver')

# Create bars for Bronze medals, with positions further offset by width
plt.bar([p + width*2 for p in positions], data['Bronze'], width, alpha=0.7, label='Bronze')

# Setting the x-axis labels to the country names, positioned in the center of the group
ax.set_xticks([p + width for p in positions])
ax.set_xticklabels(data['Country'])

# Setting the chart title and labels
plt.title('Medals Won by Country')
plt.xlabel('Country')
plt.ylabel('Medals')

# Adding the legend and showing the chart
plt.legend()
plt.tight_layout()
plt.show()
````
